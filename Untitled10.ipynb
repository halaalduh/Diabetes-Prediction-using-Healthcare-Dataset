# Load Data and Select Features

###### Load the dataset and select the features (Glucose and BMI) used for K-means clustering.


```python

#  Load data


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the preprocessed dataset
data = pd.read_csv("Preprocessed_dataset.csv")
print("Dataset loaded successfully!")
display(data.head())

# We will use 'Glucose' and 'BMI' because they are strong health indicators.
X = data[['Glucose', 'BMI']].values
print("Selected columns: Glucose and BMI")
print("X shape:", X.shape)
```

# Elbow Method (WCSS Calculation)
###### Compute WCSS for different K values to identify the "elbow point" and help decide the optimal K.


```python

# Elbow Method = WCSS


wcss = []
K_values = range(2, 11)

for k in K_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    print(f"K = {k}  ، WCSS = {kmeans.inertia_:.4f}")

plt.figure(figsize=(6, 4))
plt.plot(K_values, wcss, marker='o')
plt.title('Elbow Method (WCSS vs K)')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.grid(True, alpha=0.3)
plt.show()
```

# Silhouette Score Evaluation
###### Calculate the Silhouette Score for several K values to evaluate clustering quality and compare K choices.


```python

# Silhouette Coefficient


selected_K = [2, 3, 4, 5, 6]
silhouette_scores = []

for k in selected_K:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    labels = kmeans.fit_predict(X)
    sil_score = silhouette_score(X, labels)
    silhouette_scores.append(sil_score)
    print(f"K = {k}  ، Silhouette Score = {sil_score:.4f}")

plt.figure(figsize=(6, 4))
plt.plot(selected_K, silhouette_scores, marker='o')
plt.title('Silhouette Score for different K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid(True, alpha=0.3)
plt.show()

```

# K-Means Clustering and Visualization
###### Run K-means for K = 2, 3, and 4 and visualize the clusters and centroids on a 2D scatter plot.


```python

# applying K-Means for K = 2, 3, 4 and plot clusters


for k in [2, 3, 4]:
    # Create and fit the K-Means model
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    y_kmeans = kmeans.fit_predict(X)

    # Plot the clusters
    plt.figure(figsize=(6, 4))

    # Cluster 0
    plt.scatter(
        X[y_kmeans == 0, 0], X[y_kmeans == 0, 1],
        s=80, c='red', label='Cluster 1'
    )

    # Cluster 1
    if k > 1:
        plt.scatter(
            X[y_kmeans == 1, 0], X[y_kmeans == 1, 1],
            s=80, c='blue', label='Cluster 2'
        )

    # Cluster 2 (if exists)
    if k > 2:
        plt.scatter(
            X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s=80, c='green', label='Cluster 3'
        )

    # Cluster 3 (if exists)
    if k > 3:
        plt.scatter(
            X[y_kmeans == 3, 0], X[y_kmeans == 3, 1],
            s=80, c='purple', label='Cluster 4'
        )

    # Plot the centroids
    for i in range(k):
        plt.scatter(
            kmeans.cluster_centers_[i, 0],
            kmeans.cluster_centers_[i, 1],
            s=200, c='yellow', edgecolor='black',
            label='Centroids' if i == 0 else None
        )

    plt.title(f'K-Means Clustering (K = {k})')
    plt.xlabel('Glucose')
    plt.ylabel('BMI')
    plt.legend()
    plt.show()

    print(f"K-Means executed successfully for K = {k}")
    print("_" * 60)
```
